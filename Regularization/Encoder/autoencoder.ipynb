```python
import pandas as pd
import numpy as np

import autoencoder
import autoencoder_wn
```

```python
#Lendo a base de dados
df = pd.read_csv('ionosphere.data')
x_train = df.iloc[:, :-1].values
x_train_noisy = x_train + np.random.normal(loc=0.0, scale=0.1, size=x_train.shape)
```

```python
#Criando os objetos
aut_enc = autoencoder.Autoencoder(input_dim=x_train.shape[1], encoding_dim=14)
aut_enc_wn = autoencoder_wn.Autoencoder(input_dim=x_train.shape[1], encoding_dim=14)
```

### (a) No added noise and weight regularization

```python
# Treinamento
aut_enc.train(x_train, epochs=100, batch_size=16)
# Reconstruindo as entradas sem adição de ruído e regularização de peso
x_reconstructed = aut_enc.reconstruct(x_train)
```

```python
# Calculando o erro de reconstrução
are = np.mean(np.power(x_train - x_reconstructed, 2), axis=1)
print('Average reconstruction error(without noisy):', np.mean(are))
```
